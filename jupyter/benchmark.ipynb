```r
library(ggplot2)
library(ggpubr)
```

```r
df = data.frame()
for (ncell in c(1000,100)){
    for (transf in c('clr', 'alr')){
        for (impute in c('None', 'CZM', 'GBM', 'lrSVD', 'entropy')){
            for (i in 1:200){
                f = paste0('../results/mouseStemCells/mse/mse_', ncell, '_500_', i, '_', impute, '_', transf, '.csv')
                tmp = read.csv(f)
                tmp['ncell'] = ncell
                tmp['transf'] = transf
                tmp['impute'] = impute
                df = rbind(df, tmp)
            }
        }
    }
}
head(df)
```

```r
pos = which(df$impute == 'entropy')
df[pos,]$impute = 'freq shrink'
df$impute = factor(df$impute, levels=c('None','CZM','GBM','lrSVD','freq shrink'))
```

# Partial correlation

```r
# organize data
tmp             = df[df$corr == 'partial correlation',]
pos             = which(tmp$shrink == 'shrink std' & tmp$transf == 'clr')
tmp$shrink[pos] = 'shrink CLR'
pos             = which(tmp$shrink == 'shrink std' & tmp$transf == 'alr')
tmp$shrink[pos] = 'shrink ALR'
for (i in c('no shrinkage', 'shrink basis')){
    pos = which(tmp$shrink == i & tmp$transf == 'alr')
    tmp = tmp[-pos,]
}
tmp$shrink = factor(tmp$shrink, levels=c('no shrinkage', 'shrink CLR', 'shrink ALR', 'shrink basis'))


# N=1000
tmp1 = tmp[tmp$ncell == 1000,]
tm   = tmp1[tmp1$shrink == 'shrink ALR',]$mse
bp   = as.numeric(quantile(tm)[4]) + IQR(tm)
g1   = ggplot(tmp1, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch=T) +
        scale_y_continuous(labels = function(x) format(x, scientific = TRUE)) +
        xlab('') + 
        ylab('Mean squared error') +
        labs(fill='Imputation') +
        ggtitle('MSE partial correlation matrix, D=500, N=1000') +
        theme(
            plot.title=element_text(face="bold", size=18), 
            axis.text.x = element_text(size=18, angle=45, hjust=1), 
            axis.text.y=element_text(size=18),
            axis.title.y=element_text(size=18),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        )
                           
# N=100
tmp2  = tmp[tmp$ncell == 100,]   
tmp21 = tmp2[tmp2$shrink == 'no shrinkage',]
tmp22 = tmp2[tmp2$shrink != 'no shrinkage',]
tm    = tmp21$mse
bp    = as.numeric(quantile(tm)[4]) + IQR(tm)
g21   = ggplot(tmp21, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch=T) +
        scale_y_continuous(labels = function(x) format(x, scientific = TRUE)) +
        xlab('') + 
        ylab('Mean squared error') +
        labs(fill='Imputation') +
        theme(
            axis.text.x = element_text(size=18, angle=45, hjust=1), 
            axis.text.y=element_text(size=18),
            axis.title.y=element_text(size=18),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        ) 
tm    = tmp22[tmp22$shrink == 'shrink ALR',]$mse
bp    = as.numeric(quantile(tm)[4]) + IQR(tm)
g22   = ggplot(tmp22, aes(x=shrink, y=mse, fill=impute)) + 
        geom_boxplot(outlier.shape = NA, notch=T) +
        scale_y_continuous(labels = function(x) format(x, scientific = TRUE)) +
        xlab('') + 
        ylab('') +
        labs(fill='Imputation') +
        theme(
            axis.text.x = element_text(size=18, angle=45, hjust=1), 
            axis.text.y=element_text(size=18),
            axis.title.y=element_blank(),
            legend.text = element_text(size = 18),
            legend.title = element_text(size=18)
        ) 
g2    = ggarrange(g21, g22, widths=c(0.55,1), legend='none')       
g2    = annotate_figure(g2, top = text_grob('MSE partial correlation matrix, D=500, N=100', color = "black", face = "bold", size = 18))

                        
                           
# plot figure
options(repr.plot.width=15, repr.plot.height=10)                           
g = ggarrange(g1, NULL, g2, nrow=1, ncol=3, legend='bottom', common.legend=T, widths=c(1,.1,1))
g
```
